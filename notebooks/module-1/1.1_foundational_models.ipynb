### DeepSeek (`langchain-deepseek`)

`ChatDeepSeek` is a dedicated LangChain integration for DeepSeek's hosted API — not a workaround via `ChatOpenAI` with a custom `base_url`. It reads `DEEPSEEK_API_KEY` from the environment automatically.

Two models are available:

| Model string | Maps to | Tool calling | Structured output |
|---|---|---|---|
| `deepseek-chat` | DeepSeek-V3 | ✅ | ✅ |
| `deepseek-reasoner` | DeepSeek-R1 | ❌ | ❌ |

Use `deepseek-chat` for agent notebooks. `deepseek-reasoner` is a thinking/reasoning model and does not support tool calling or structured output.

The `invoke`, `stream`, and `with_structured_output` interface is identical to all other LangChain providers — that is the point of LangChain's abstraction.

Docs: https://python.langchain.com/docs/integrations/chat/deepseek/

```python
from langchain_deepseek import ChatDeepSeek

model = ChatDeepSeek(model="deepseek-chat")

response = model.invoke("What's the capital of the Moon?")
print(response.content)
```

## Initialising and invoking an agent

```python
from langchain.agents import create_agent

agent = create_agent(model=model)
```

```python
from langchain.messages import HumanMessage

response = agent.invoke(
    {"messages": [HumanMessage(content="What's the capital of the Moon?")]}
)
```

```python
from pprint import pprint

pprint(response)
```

```python
print(response['messages'][-1].content)
```

```python
from langchain.messages import AIMessage

response = agent.invoke(
    {"messages": [HumanMessage(content="What's the capital of the Moon?"),
    AIMessage(content="The capital of the Moon is Luna City."),
    HumanMessage(content="Interesting, tell me more about Luna City")]}
)

pprint(response)
```

## Streaming Output

```python
for token, metadata in agent.stream(
    {"messages": [HumanMessage(content="Tell me all about Luna City, the capital of the Moon")]},
    stream_mode="messages"
):

    # token is a message chunk with token content
    # metadata contains which node produced the token
    
    if token.content:  # Check if there's actual content
        print(token.content, end="", flush=True)  # Print token
```
